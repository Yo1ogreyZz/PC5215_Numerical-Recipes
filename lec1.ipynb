```python
def fibonacci_generator():
    n = int(input("Enter the number of Fibonacci: "))

    # Handle edge cases
    if n < 0:
        print("Please enter a non-negative number")
        return

    # Initialize the Fibonacci sequence
    if n == 0:
        F = [0]
    else:
        F = [0, 1]
        # Generate remaining Fibonacci numbers
        for i in range(2, n + 1):
            F.append(F[i-1] + F[i-2])

    # Print Fibonacci numbers
    for i in range(n + 1):
        print(f"Fibonacci({i}) = {F[i]}")

    # Print ratio F(n-1)/F(n) only if n > 0 and F[n] != 0
    if n > 0 and F[n] != 0:
        ratio = F[n-1] / F[n]
        print(f"The ratio F_{n-1}/F_{n} = {ratio}")
    elif n == 0:
        print("Ratio is undefined for n = 0")

# Call the function
fibonacci_generator()
```

```python
import sys
import numpy as np

print("=== 64位双精度浮点数特性 ===")

# 方法1：使用sys模块
print(f"机器精度 (sys.float_info.epsilon): {sys.float_info.epsilon}")
print(f"最大浮点数 (sys.float_info.max): {sys.float_info.max}")

# 方法2：使用numpy
print(f"机器精度 (numpy): {np.finfo(float).eps}")
print(f"最大浮点数 (numpy): {np.finfo(float).max}")

# 验证机器精度的定义
eps = sys.float_info.epsilon
print(f"\n=== 验证机器精度定义 ===")
print(f"1.0 + eps/2 == 1.0: {1.0 + eps/2 == 1.0}")
print(f"1.0 + eps == 1.0: {1.0 + eps == 1.0}")
print(f"机器精度是使得 1.0 + ε > 1.0 成立的最小正数")

# 理论值验证
theoretical_eps = 2**(-52)
theoretical_max = (2 - 2**(-52)) * 2**1023
print(f"\n=== 理论值对比 ===")
print(f"理论机器精度 2^(-52): {theoretical_eps}")
print(f"实际机器精度: {eps}")
print(f"是否匹配: {abs(theoretical_eps - eps) < 1e-20}")

print(f"\n理论最大值: {theoretical_max}")
print(f"实际最大值: {sys.float_info.max}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def unstable_recursion(n, F0=1, F1=1):
    """使用不稳定的递推关系计算"""
    F = [F0, F1]
    for i in range(2, n+1):
        F.append(F[i-2] - F[i-1])
    return F

def golden_ratio_power(n):
    """理论值：黄金分割比的幂"""
    phi = (1 + np.sqrt(5)) / 2
    return phi ** n

# 测试不稳定性
n_max = 30
F_unstable = unstable_recursion(n_max)
theoretical = [golden_ratio_power(i) for i in range(n_max+1)]

print("=== 数值不稳定性演示 ===")
print("n\t递推计算\t理论值\t\t相对误差")
print("-" * 50)

for i in range(0, n_max+1, 5):
    if i < len(F_unstable):
        rel_error = abs(F_unstable[i] - theoretical[i]) / theoretical[i]
        print(f"{i}\t{F_unstable[i]:.6e}\t{theoretical[i]:.6e}\t{rel_error:.6e}")

# 可视化误差增长
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
n_vals = range(min(20, len(F_unstable)))
plt.plot(n_vals, [F_unstable[i] for i in n_vals], 'ro-', label='递推计算')
plt.plot(n_vals, [theoretical[i] for i in n_vals], 'b--', label='理论值')
plt.xlabel('n')
plt.ylabel('F_n')
plt.title('递推计算 vs 理论值')
plt.legend()
plt.yscale('log')

plt.subplot(1, 2, 2)
rel_errors = [abs(F_unstable[i] - theoretical[i]) / theoretical[i]
              for i in range(min(20, len(F_unstable)))]
plt.plot(range(len(rel_errors)), rel_errors, 'g.-')
plt.xlabel('n')
plt.ylabel('相对误差')
plt.title('相对误差随n的增长')
plt.yscale('log')

plt.tight_layout()
plt.show()
```
